```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file into a pandas DataFrame
data = pd.read_csv('/content/medical.csv', encoding='utf-8')

# Drop rows with missing 'Age' values
data.dropna(subset=['Age'], inplace=True)

# Convert 'Age' column to numeric (if needed)
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')

# Filter out non-numeric values (if needed)
data = data[data['Age'].notna()]

# Define age group bins and labels
age_bins = [0, 20, 30, 40, 50, 60, 70, 80, 90]
age_labels = ['1-20', '20-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90']

# Categorize age into different age groups
data['Age Group'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels, right=False)

# Count the number of patients in each age group
age_group_counts = data['Age Group'].value_counts()

# Plotting the pie chart for age distribution
plt.figure(figsize=(8, 8))
plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Age Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.

# Adding legend for age group labels
plt.legend(age_labels, title='Age Group', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file into a pandas DataFrame
data = pd.read_csv('/content/medical.csv')

# Drop rows with missing 'Age' values
data.dropna(subset=['Age'], inplace=True)

# Convert 'Age' column to numeric (if needed)
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')

# Filter out non-numeric values (if needed)
data = data[data['Age'].notna()]

# Define age group bins and labels
age_bins = [0, 20, 30, 40, 50, 60, 70, 80, 90]
age_labels = ['<20', '20-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90']

# Categorize age into different age groups
data['Age Group'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels, right=False)

# Group data by Age Group and Sex, and count occurrences
grouped_data = data.groupby(['Age Group', 'Sex']).size().unstack()

# Print the grouped_data to check the column names
print(grouped_data)

# Plotting the bar chart for the classification of sex based on age group
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(age_labels))

# Update the column names based on the actual column names in grouped_data
plt.bar(index, grouped_data['M'], bar_width, label='Male', color='blue')
plt.bar(index, grouped_data['F'], bar_width, bottom=grouped_data['M'], label='Female', color='pink')

plt.xlabel('Age Group')
plt.ylabel('Number of Patients')
plt.title('Classification of Sex based on Age Group')
plt.xticks(index, age_labels)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file into a pandas DataFrame
data = pd.read_csv('/content/medical.csv')

# Convert 'Age' column to numeric and replace non-numeric values with NaN
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')

# Define age group bins
age_bins = [0, 20, 30, 40, 50, 60, 70, 80, 90]

# Categorize age into different age groups
data['Age Group'] = pd.cut(data['Age'], bins=age_bins, right=False)

# Convert 'Age Group' column back to strings
data['Age Group'] = data['Age Group'].astype(str)

# Filter the rows where 'HVKP In PTA' column is 'Positive'
hvkp_positive_data = data[data['HVKP In PTA'] == 'P']

# Drop rows with missing 'Sex' values
hvkp_positive_data.dropna(subset=['Sex'], inplace=True)

# Get all unique values in the 'Sex' column and set 'F' and 'M' explicitly
sex_labels = ['F', 'M']
hvkp_positive_data.loc[:, 'Sex'] = pd.Categorical(hvkp_positive_data['Sex'], categories=sex_labels)

# Group the filtered data by 'Sex' and 'Age Group', and count occurrences
grouped_data = hvkp_positive_data.groupby(['Sex', 'Age Group']).size().unstack()

# Plotting the stacked bar chart
grouped_data.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10, 6))
plt.xlabel('Age Group')
plt.ylabel('Number of HvKp Positive Cases')
plt.title('HvKp Positive Total with Respect to Sex and Age')
plt.legend(title='Sex', loc='upper right', labels=['Female', 'Male'])
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
